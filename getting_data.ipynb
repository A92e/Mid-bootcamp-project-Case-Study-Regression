```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from time import sleep
import numpy as np
```

```python

##########       GET ALL THE LINKS ONE PAGE        ############

base_url = 'https://www.pisos.com/venta/pisos-valencia_capital/desde-120000/hasta-210000/'

# request and create bs
pisos = requests.get(base_url)
soup = BeautifulSoup(pisos.content, 'html.parser')

num_iter = len(soup.select('div[itemtype="http://schema.org/SingleFamilyResidence"]'))

urls_page = []
urls_page.append(base_url)
#create link pages
for i in range(2, 76):
    urls_page.append(base_url+ str(i) +'/')



# ### Create urls list for each add per page
def getting_url(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')

    num_iter = len(soup.select('div[itemtype="http://schema.org/SingleFamilyResidence"]'))

    urls = []
    for i in range(num_iter):
        ref = soup.select('div[itemtype="http://schema.org/SingleFamilyResidence"]')[i].get('data-navigate-ref')
        base = 'https://www.pisos.com'
        url = base+ref
        urls.append(url)

    return urls


house_urls = []
for url in urls_page:
    print(url)
    house_urls.append(getting_url(url))
    wait_time = np.random.randint(1, 4)
    sleep(wait_time)
```

```python
# list of house urls
urls = []
for i in house_urls:
    for x in i:
        urls.append(x)
```

```python
##### Create a funtion that takes the wanted elements for each url
def get_info(url):
    house = requests.get(url)
    soup = soup = BeautifulSoup(house.content, 'html.parser')
    
# create lists to store info
    sqft_build = []
    sqft_living = []
    bedrooms = []
    bathrooms = []
    floor = []
    view = []
    grade = []
    latitude = []
    longitude = []
    district = []
    
# append info to lists
# # sqft_build
    try:
        sqft_build.append(soup.select_one('li[class="charblock-element more-padding"]').select_one('span[class="icon icon-inline icon-default icon-superficieconstruida"]')
                    .parent.select('span')[1].get_text().split(' ')[1])
    except:
        sqft_build.append('None')

# # # sqft_living
    try:
        sqft_living.append(soup.select_one('div[id="characteristics"]').select_one('span[class="icon icon-inline icon-default icon-superficieutil"]')
                .parent.select('span')[1].get_text().split(' ')[1])
    except:
        sqft_living.append('None')
        
# # # rooms
    try:
        bedrooms.append(soup.select_one('div[id="characteristics"]').select_one('span[class="icon icon-inline icon-default icon-numhabitaciones"]')
                .parent.select('span')[1].get_text().strip(':').strip(' '))
    except:
        bedrooms.append('None')
        
# # # bathrooms
    try:
        bathrooms.append(soup.select_one('div[id="characteristics"]').select_one('span[class="icon icon-inline icon-default icon-numbanos"]')
                .parent.select('span')[1].get_text().strip(':').strip(' '))
    except:
        bathrooms.append('None')
# # # floor     
    try:
        floor.append(soup.select_one('div[id="characteristics"]').select_one('span[class="icon icon-inline icon-default icon-planta"]')
                .parent.select('span')[1].get_text().strip(':').strip(' ').strip('ª').strip('º'))
    except:
        floor.append('None')

# # # view
    try:
        view.append(soup.select_one('div[id="characteristics"]').select_one('span[class="icon icon-inline icon-default icon-exterior"]')
                .get_text().strip(':').strip(' '))
    except:
        view.append('None')

# # # grade
    try:
        grade.append(soup.select_one('div[id="characteristics"]').select_one('span[class="icon icon-inline icon-default icon-estadoconservacion"]')
                .parent.select('span')[1].get_text().strip(':').strip(' '))
    except:
        grade.append('None')
    
# # # district
    try:
        district.append(soup.select_one('div[class="maindata-info"]').select('h2')[0].get_text().split('(')[0].strip(' '))
    except:
        pass
# create a dictionary to store all the lists
    info = {'sqft_build': sqft_build,
            'sqft_living': sqft_living,
            'bedrooms': bedrooms,
            'bathrooms': bathrooms,
            'floor':  floor,
            'view': view,
            'grade': grade,
            'district': district
           }
    
    return info
```
